```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('happyscore_income.csv')
```

```python

top_gdp = data[ data['GDP'] > 1.2 ] #filter

top_gdp.sort_values('GDP', ascending="True", inplace=True) #Sorting the values in the data frame by GDP (Gross Domestic Product)
```

```python
plt.scatter(top_gdp['GDP'],top_gdp['avg_income'])

plt.xlabel("GDP")
plt.ylabel("avg_income")
plt.legend(loc='upper left')

plt.text(top_gdp.iloc[0]['GDP'],
         top_gdp.iloc[0]['avg_income'],
         top_gdp.iloc[0]['country'])

plt.text(top_gdp.iloc[-1]['GDP'],
         top_gdp.iloc[-1]['avg_income'],
         top_gdp.iloc[-1]['country'])
```
